# Genome Databases Overview

## Reference Genome Resources

### NCBI / RefSeq
- Main global archive of DNA sequences across all life forms  
- Includes bacteria, viruses, plants, animals, and humans  
- Each organism has a standard reference genome version (RefSeq)  

---

### Ensembl (EMBL-EBI)
- Focuses mainly on animals (especially vertebrates)  
- Provides biological interpretation, not just raw sequences  
- Enables side-by-side comparison between species  
- Useful for identifying:
  - Orthologs: same gene in different species  
  - Paralogs: duplicated genes within a species  

---

### UCSC Genome Browser
- Visual tool for exploring genomes like a map  
- Displays genes and features as stacked tracks  
- Used for visualizing:
  - genes  
  - mutations  
  - repeats  
  - regulatory regions  
- Allows uploading your own sequencing data and comparing across species  

---

## Variant Databases

### dbSNP (NCBI)
- Stores small DNA variants:
  - SNPs (single nucleotide polymorphisms)  
  - small insertions/deletions (indels)  
- Used to check whether a variant is common or rare  

---

### dbVAR (NCBI)
- Stores structural variants:
  - deletions  
  - duplications  
  - inversions  
  - rearrangements  
- Used for analyzing large-scale genomic variation 

---

### ClinVar (NCBI)
- Focuses on clinical interpretation of variants  
- Classifies variants as:
  - harmless  
  - disease-causing  
  - uncertain significance  
- Used for:
  - diagnosing genetic diseases  
  - interpreting patient sequencing results  

---

### EVA (European Variation Archive)
- Accepts variant data from any organism 
- Used for:
  - sharing  
  - storing large-scale variant datasets  

---

## Key Takeaway
Different genome resources are useful depending on the type of data you want to analyze:
- Reference genomes → RefSeq  
- Comparative genomics → Ensembl  
- Visualization → UCSC  
- Variant analysis → dbSNP, dbVAR, ClinVar, EVA
