# Clinical Coding and Survival Analysis Notes

## ICD-10
Universal labeling system for diagnoses.  
Each disease gets a code made of letters and numbers.  

Example:  
C91.1  
- C = cancer  
- 91 = leukemia  
- .1 = specific type  

---

## ICD-O-3
Uses two codes:
- morphology (what the cancer cell looks like under a microscope, e.g. gland-forming cancer)  
- topology (where in the body it is, e.g. lung)  

You always need both pieces of information.  

Weakness:
- no hierarchy (cannot easily find subtypes)  
- searching for "adenocarcinoma" returns only that exact code  
- cannot automatically retrieve all subtypes  
- must manually know and search each subtype code  

---

## NCIt
- One single code per cancer type  
- Hierarchical system  
- No need to know all subtype codes in advance

## Kaplan–Meier plots
The Kaplan–Meier estimator tracks how many people remain event-free over time.  
(Event is often death, but can be any defined outcome.)  

### How the curve works
- Start with a group of patients and follow them over time  
- Each time someone has the event (e.g. dies) → the curve goes down  
- Curve starts at 100% (everyone alive/event-free)  
- Decreases in steps, not smoothly  

### Interpretation
- Drop in line = an event happened  
- Tick mark (no drop) = censoring  
  - person left the study or was lost  
  - event did not happen to them  
- Curve represents step-by-step survival over time  

### Key idea
At each step:  
- out of the people still being followed, how many survived that moment  
- this is conditional probability  

### Risk groups
- Patients can be split into groups based on risk  
- Each group has its own survival curve  

Example:
- del(17p) marker → patients tend to do worse  
- their survival curve drops faster  

### Statistics
- log-rank test → checks if differences between curves are significant  

### Clinical use
- measure patient markers  
- assign to a risk group  
- estimate probability of being event-free at 2, 5, or 10 years

## TNM codes

### TNM staging
Describes how far a tumor has spread  

- T = tumor size/extent  
  - T1 = small  
  - T4 = large/invasive  

- N = lymph node involvement  
  - N0 = no  
  - N1/N2/N3 = increasing spread  

- M = metastasis (spread to other organs)  
  - M0 = no  
  - M1 = yes  

---

## Examples

T2N0M0  
- medium-sized tumor  
- no lymph node involvement  
- no distant metastasis  

T3N2M0  
- large tumor  
- significant lymph node involvement  
- no spread to other organs  

T4N3M1  
- very large/invasive tumor  
- extensive lymph node involvement  
- distant metastases present  

T1N0M0  
- small tumor  
- no lymph node involvement  
- no metastasis
