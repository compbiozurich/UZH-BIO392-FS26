Download and interpret CNV frequency data

Now download the CNV frequency table for:

Small cell carcinoma, NOS

Tasks:

download the frequency table

what one row represents in the frequency table
why these two tables cannot be merged row by row

```python
import requests
import pandas as pd
from io import StringIO

freq_url = {"Small_cell_carcinoma": "https://progenetix.org/services/intervalFrequencies/?output=pgxseg&filters=pgx:icdom-80413"}

for query, url in freq_url.items():
    print(f"Downloading {query}")
    response = requests.get(url, timeout=120)
    response.raise_for_status()
    df_freq = pd.read_csv(StringIO(response.text), sep="\t", comment="#")
    df_freq["source_query"] = query

# inspect shape
print(df_freq.shape)

# inspect columns
print(df_freq.columns)
```

- The frequecy table is __aggregated__ data. This means that instead of displaying information about individual samples, it contains a summarized form of the data, which appears in the Sample table.
- This inherently means, that their shapes will differ, and the information within teh table does not mean the same thing! Hence they can NOT be merged.
- A row in the frequency table corresponds to a genomic region (bin)  
- A row in a sample table, corresponds to a single sample (e.g. individual, tumor, ...)
